```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.ticker as ticker
from mpl_toolkits.axes_grid1 import ImageGrid
import pdb
import pickle as pickle

from data.source import *
from stochastic import *
from streaming import *
from static import *
from adaptive import *
```

```python
source = NumpyFileSource('yale_faces.npz')
sigma = np.std(source.train_data)

minibatch_size = 10

alpha = 5.0
sigma_a = sigma * 0.5
sigma_x = sigma * 0.5

t0 = 1
kappa = 0.5

truncation = 100

stochastic_model = StochasticInfinite(
    dim=source.dim,
    data_size=source.size,
    num_features=truncation,
    alpha=alpha,
    sigma_a=sigma_a,
    sigma_x=sigma_x,
    t0=t0,
    kappa=kappa,
)
stochastic_strategy = Static(stochastic_model, source, minibatch_size)
```

```python
streaming_model = StreamingInfinite(
    dim=source.dim,
    num_features=truncation,
    alpha=alpha,
    sigma_a=sigma_a,
    sigma_x=sigma_x,
    t0=t0,
    kappa=kappa,
)
streaming_strategy = Static(streaming_model, source, minibatch_size)
```

```python
streaming_adaptive_model_kwargs = {
    'dim': source.dim,
    'prior_truncation': truncation,
    'alpha': alpha,
    'sigma_a': sigma_a,
    'sigma_x': sigma_x,
    't0': t0,
    'kappa': kappa,
}
streaming_adaptive_strategy = Adaptive(streaming_adaptive_model_kwargs, source, minibatch_size, elbo_check_interval=1)
```

```python
strategies = [
    stochastic_strategy,
    streaming_strategy,
    streaming_adaptive_strategy,
]
strategy_names = [
    'Stochastic',
    'Streaming',
    'Streaming-Adaptive',
]
```

```python
for i, strategy in enumerate(strategies):
    source.reset_cursor()

    while source.has_next():
        if strategy.iteration % 10 == 0:
            print(strategy_names[i], strategy.iteration)

        strategy.loop()
```

```python
# Plot features
width = 32
height = 32


def plot_feature(ax, feature, vmin, vmax):
    im = ax.imshow(feature.reshape(width, height).T,
                   interpolation='nearest',
                   cmap=cm.RdBu_r,
                   vmin=vmin,
                   vmax=vmax,)

    ax.xaxis.set_major_locator(ticker.NullLocator())
    ax.yaxis.set_major_locator(ticker.NullLocator())

    return im

for i, strategy in enumerate(strategies):
    f = plt.figure()

    features = strategy.features()
    num_features = len(features)

    num_rows = num_features // 10
    if num_features % 10 > 0:
        num_rows += 1

    grid = ImageGrid(f, 111,
                     nrows_ncols=(num_rows, 10),
                     axes_pad=0,
                     cbar_mode='single',
                     cbar_location='right',
                     cbar_pad=0.05)
    cax = grid.cbar_axes[0]

    vmax = np.abs(features).max()
    vmin = -vmax

    for k, feature in enumerate(features):
        im = plot_feature(grid[k], feature, vmin, vmax)

    for k in range(num_features, num_rows * 10):
        grid[k].set_visible(False)

    cax.colorbar(im)

# Plot predictive log-likelihood against amount of data seen
f = plt.figure()
ax = f.add_subplot(111)

handles = []
for i, strategy in enumerate(strategies):
    handle = ax.plot(strategy.data_seen_set, strategy.ll_mean_set,
                     linewidth=2.0,
                     label=strategy_names[i])[0]
    handles.append(handle)

ax.legend(handles=handles, labels=strategy_names, loc='lower right')
ax.set_ylabel('Predictive Log-Likelihood')
ax.set_xlabel('Amount of Data Seen')

# Plot predictive log-likelihood against time
f = plt.figure()
ax = f.add_subplot(111)

handles = []
for i, strategy in enumerate(strategies):
    handle = ax.errorbar(strategy.time_set, strategy.ll_mean_set,
                         linewidth=2.0,
                         label=strategy_names[i])[0]
    handles.append(handle)

ax.legend(handles=handles, labels=strategy_names, loc='lower right')
ax.set_ylabel('Predictive Log-Likelihood')
ax.set_xlabel('Time (s)')

# Plot parameter tying level against amount of data seen
f = plt.figure()
ax = f.add_subplot(111)
ax.plot(streaming_adaptive_strategy.data_seen_set, streaming_adaptive_strategy.truncation_set,
        linewidth=2.0)
ax.set_ylim((0, truncation + 1))
ax.set_ylabel('Parameter Tying Level')
ax.set_xlabel('Amount of Data Seen')

# Show plots
plt.show(block=False)

# Print performance results
for i, strategy in enumerate(strategies):
    print(strategy_names[i], strategy.time_set[-1], strategy.ll_mean_set[-1])

# Save results
pickle.dump(strategies, open('strategies.pkl', 'wb'))

# Anything else to explore?
pdb.set_trace()
```
